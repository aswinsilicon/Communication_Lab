```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 65
var = 1
N = 100000
```

```python
X = np.random.normal(mean, var, N)
Y = np.random.normal(mean, var, N)

print("X mean:", np.mean(X), ", var:", np.var(X))
print("Y mean:", np.mean(Y), ", var:", np.var(Y))
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Histogram of X")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.hist(X, bins=30, density=False, color='orange')
plt.subplot(1, 2, 2)
plt.title("Histogram of Y")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.hist(Y, bins=30, density=False, color='red')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.title("PDF of X")
plt.xlabel("Bins")
plt.ylabel("Relative Frequency")
plt.hist(X, bins=30, density=True, color='orange')
plt.subplot(1, 2, 2)
plt.title("PDF of Y")
plt.xlabel("Bins")
plt.ylabel("Relative frequency")
plt.hist(Y, bins=30, density=True, color='red')
plt.show()
```

```python
Z = X + Y
plt.hist(Z, bins=30, density=True)
plt.title("Histogram of Z")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
Z_mean = np.mean(Z)
Z_var = np.var(Z)

Z_meant = 2 * mean
Z_vart = 2 * var

print("Mean of Z:", Z_mean, ", Theoretical:", Z_meant, ", Error:", Z_meant - Z_mean)
print("Var of Z:", Z_var, ", Theoretical:", 2 * var, ", Error:", Z_vart - Z_var)
```

```python
p1 = np.sum(np.abs(X-mean) <= 1 * var) / N
p2 = np.sum(np.abs(X-mean) <= 2 * var) / N
p3 = np.sum(np.abs(X-mean) <= 3 * var) / N

print("68% Prob mass:", p1, ", Error:", np.abs(p1-0.68))
print("95% Prob mass:", p2, ", Error:", np.abs(p2-0.95))
print("99.7% Prob mass:", p3, ", Error:", np.abs(p3-0.997))
```
